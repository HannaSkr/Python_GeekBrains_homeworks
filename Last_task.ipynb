```python
#решение задачи без использования get_dummies
#https://dfedorov.spb.ru/pandas/ - использовала для вдохновения
import pandas as pd
import numpy as np
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data["one_hot_code"] = np.where(data["whoAmI"].str.contains("human"), 1, 0)
data.head(10)
```

```python
#решение задачи с get_dummies
#https://www.codecamp.ru/blog/pandas-get-dummies/
import pandas as pd
import numpy as np
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
pd.get_dummies(data, columns=["whoAmI"], drop_first=True).head(10)
```
